```python
#apskaiciuokite kiekvieno studento vidurki ir prie vidurkio nurodykite 
#jei vidurkisprint("Studentai:")
#isveskite atskirose eilutese graziai suformatuota informacija apie studentus
studentai = []

with open('atsiskaitymas_studentai.txt', encoding="utf8") as failas:
    next(failas)
    for eilute in failas:
        eilute = eilute.rstrip('\n')  
        isskaidyta = eilute.split(',') 
        pazymiai = [int(pazymys) for pazymys in isskaidyta[6:]]
        vidurkis = sum(pazymiai) / len(pazymiai) 
        leidziama = "Leidziama laikyti egzamina" if vidurkis >= 5 else "Neleidziama laikyti egzamino"

        studentas = [
            isskaidyta[0],  
            isskaidyta[1],  
            int(isskaidyta[2]),  
            int(isskaidyta[3]),  
            isskaidyta[4], 
            isskaidyta[5],  
            pazymiai, 
            round(vidurkis, 2), 
            leidziama  
        ]
        studentai.append(studentas)


for index, studentas in enumerate(studentai, start=1):
    print(f"{index}. {studentas[0]} {studentas[1]}, Vidurkis: {studentas[7]}, {studentas[8]}")
    
#spausdiname i faila atsiskaitymas_rezultatai
with open('atsiskaitymo_rezultatai.txt', 'w', encoding="utf8") as rezultatai_failas:
    for index, studentas in enumerate(studentai, start=1):
        rezultatai_failas.write(f"{index}. {studentas[0]} {studentas[1]}\n")
        rezultatai_failas.write(f"   Amzius: {studentas[2]}\n")
        rezultatai_failas.write(f"   Ugis: {studentas[3]} cm\n")
        rezultatai_failas.write(f"   Miestas: {studentas[4]}\n")
        rezultatai_failas.write(f"   Mokymosi staiga: {studentas[5]}\n")
        rezultatai_failas.write(f"   Pazymiai: {studentas[6]}\n")
        rezultatai_failas.write(f"   Vidurkis: {studentas[7]}\n")
        rezultatai_failas.write(f"   {studentas[8]}\n\n")
        


    
```
